# DATA EXTRACTION

```python
import zipfile
import os

# Define the ZIP file name and extraction path
zip_path = 'datasets.zip'
extract_path = './'

# Create the destination folder if it doesn't exist
os.makedirs(extract_path, exist_ok=True)

# Extract the ZIP file contents
with zipfile.ZipFile(zip_path, 'r') as zip_ref:
    zip_ref.extractall(extract_path)

print(f"✅ Dataset successfully extracted")
```

# BASIC INFO AND STATISTICS

```python
import pandas as pd

# Load the dataset
df = pd.read_csv('datasets/test_0_rows_500_columns_20_comp_0.2_input_.csv')

# Display shape
print("Dataset shape:", df.shape)

# Show basic info
print("\nData types and non-null counts:")
df.info()

# Show percentage of missing values per column
missing_percentage = df.isnull().mean() * 100
print("\nPercentage of missing values per column:")
print(missing_percentage.sort_values(ascending=False))

# Show basic statistics
print("\nDescriptive statistics:")
print(df.describe())

# Display first few rows
df.head()
```
